# 1. Statistical Learning

#### Stastical Learning
data를 통해서 pattern을 찾고, 해석가능한 형태의 모델을 구축함으로써 예측(회귀)이나 분류 문제 등을 풀어내는 통계적인 방법 


#### 변수를 가지고 모델 만드는 방법
- Parametric model: 모델의 형태를 가정하고, 데이터를 통해(학습을 통해) 각 파라미터를 추정하여 모델을 완성(lr, LDA, Naive Bayes, 단순 신경망)
- Non-Parametric model: 모델의 형태가 어떨 것이라는 것을 미리 가정하지 않고 추정하는 방법(KNN, Decision Tree, Boosting/Random Forest, SVM, 복잡 신경망) 

#### 오차 개념
- 모델은 어쩔 수 없이 줄일 수 없는 오차가 있다(Irreducible error)

- 이러한 모델의 오차를 통해서 모델을 평가하고, 어떤 모델을 선택할지 판단이 필요하다 
#### bias-variance trade off 개념(underfit-overfit)
- bias: 모델의 예측값과 실제값과 차이.bias가 크다는 것은 모델이 실제값을 잘 예측하지 못하고 있다는 의미임(underfit) -> 추정치와 실제값을 이용해 계산   
  - **High bias(underfit): 채택한 모델의 flexibility가 낮아서(학습력이 부족해) X, Y 간의 복잡한 관계를 충분히 표현할 수 없는 경우. 훈련용 데이터(test set)이 많아도 학습이 충분히 되지 않으며 X로 Y를 잘 맞추지 못한다.**
- variance: 다른 training data set을 이용해 다시 모델을 학습시킬 때 그 예측한(추정된) 결과값의 변화량을 의미함. model이 특정 dataset에 대해 overfit 되었다면 다른 dataset에 대한 결과는 이전의 결과와 상이할 것이다. variance가 높다는 것은 overfit 되었다는 의미임. -> 추정치만 이용해서 잔차 계산 
  - **High Variance(overfit): training set 데이터에 따라 너무 민감하게 반응함. 학습력이 지나쳐서 training set의 random noise에도 반응해 아예 train set을 외우는 특성을 보인다. 이런 모델은 실제 데이터에 적용하면 성능이 안좋다.**

- Typically as the flexibility of f hat increases, its variance increases, and its bias decreases. So choosing the flexibility based on average test error amounts to bias-variance trade-off.                    
- **Bias, Variance 두 에러를 동시에 낮추는 모델이 바람직함. 하지만 하나를 낮추는 모델은 다른 하나를 높이는 경향을 보인다.**
