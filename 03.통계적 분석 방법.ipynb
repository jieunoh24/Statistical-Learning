# 3. 통계적 분석 방법
---

- **상관관계분석(Correlation Analysis)**               
연구하고자 하는 변수들간의 관련성을 분석하기 위해 사용함. 한 변수가 다른 변수와의 관련성이 있는지 여부와 관련성이 있다면 어느정도의 관련성이 있는지를 알고자 할때 이용함. 상관관계분석은 각각의 변수가 주로 연속형 데이터인 경우에 사용함. 
- **요인분석(Factor Analysis)**              
일련의 관측된 변수에 근거하여 직접 관측할 수 없는 요인을 확인하기 위한 것임. 수많은 변수들을 적은 수의 몇가지 요인으로 묶어줌으로써 그 내용을 단순화하는 것이 목적임.
- **회귀분석(Regression Analysis)**                    
다변량모집단에서는 모집단을 구성하는 변수들 간에 어떠한 관계가 있는가? 그리고 변수들 간의 인과관계가 존재한다면 이를 어떻게 함수로 표시하는 것이 합리적인가에 대한 문제를 주로 다루고 있음. 회귀분석은 변수들 중 하나를 종속변수로 나머지를 독립변수로 하여 이들 변수들이 서로 상관관계를 가질 때 독립변수가 변화함에 따라 종속변수가 어떻게 변화하는가를 규명하는 통계기법임. 독립변수의 개수에 따라 단순회귀분석과 다중회귀분석으로 구분할 수 있음. 
- **T-test 분석**           
일반적으로 두 집단의 평균을 비교하는 분석방법에는 크게 Z-검정과 T-검정으로 구분되는데 Z-검정은 모집단의 분산을 알고 있는 경우에 사용됨. 그러나 두 모집단의 분산을 알고 있는 경우는 드물기 때문에 보통 T-검정을 사용하게 됨. T-검정은 두 집단간의 평균의 차이가 통계적으로 유의한지를 파악할 때 이용하는 통계기법임. 
- **분산분석(Analysis of Variance: ANOVA)**         
두 표본 이상의 평균치에 대한 차이를 검정하는 통계기법임. 분산분석을 이용하여 표본들이 동일한 평균을 가진 모집단에서 추출된 것인지의 여부를 추론할 수 있음.
- **판별분석(Discriminants Analysis)**        
사회현상의 여러 특성들을 토대로하여 주어진 상황에서 응답자들이 어떻게 행동할 것인지를 예측하는 하나의 통계기법임.
- **군집분석(Cluster Analysis)**      
분류할 집단에 특정한 대상물을 배정하여 동일 집단의 대상물이 유사성을 갖게 함으로써 집단간의 차이를 명확하게 하는데 이용됨. 군집분석의 목적은 동질적인 집단이나 군집을 분류하는 데 있음.
