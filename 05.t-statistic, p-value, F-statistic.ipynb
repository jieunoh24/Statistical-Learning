# 5. t-statistic, p-value, F-statistic
---
- null hypothesis(귀무가설)
- X와 Y는 관계가 없다는 가설임 

- 이 귀무가설을 채택할 것인지 기각할 것인지에 대한 기준을 t-통계량을 활용함
- t-통계량은 쉽게 말하면 [추정된 파라미터값/표준편차(파라미터값)]으로 볼 수 있음

- t-통계량이 클수록 p-value는 작아지고, 귀무가설을 기각할 확률이 높아짐
- t-통계량이 작을수록 귀무가설을 채택할 확률이 높아져 X와 Y의 관계가 없음을 통계적으로 확인할 수 있음

- F-통계량은 다중의 X와 Y간 관계를 파악할 때 활용됨 

- T-test를 하게 되면 T, P 값이, ANOVA를 하게 되면 F, P 값이, 회귀분석에서는 F, T, P 값이 교차분석에서는 카이제곱, P 값이 나옴
